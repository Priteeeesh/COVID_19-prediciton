```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("covid_19_india.csv")
```

```python
df.Date=pd.to_datetime(df.Date)
df['Month']=df.Date.dt.month
df.drop('Sno',inplace=True,axis=1)
```

```python
dff = df.groupby('Date').sum().sort_values(by=['Confirmed'])
dff.head(50)
```

```python
dff=dff.drop(['Month','Cured'],axis=1)
dff.tail()
```

```python
import matplotlib.pyplot as plt
plt.plot(dff)
plt.show()
```

```python
dff.to_csv("Confirmed.csv")
```
